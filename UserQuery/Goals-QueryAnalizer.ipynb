<h1>Goal Modelling (GoMo)</h1>
<br />
A <i>goal model</i> is a type of statistical model for discovering the abstract "goals" that occur in a collection of queries. The algorithms in GoMo, such as <b>Latent Dirichlet Allocation (LDA)</b>, discover semantic structure of queries, by examining word statistical co-occurrence patterns within a corpus of training <b>explicit intentional queries (EIQ)</b>. These algorithms are unsupervised, which means no human input is necessary – you only need a corpus of plain text EIQ.
