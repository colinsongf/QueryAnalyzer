<h1>Calculating LDA perplexit</h1>

```python
import numpy
import scipy
import gensim

import random
#random.seed(11091987)           #set random seed

import ldamodel as ldam
import generic_gomo as g


model_path = "/media/University/UniversityDisc/2-Master/MasterThesis/EjecucionTesis/Desarrollo/PythonProjects/QueryAnalyzer/Models/"
corpus, dictionary = g.loadCorpusDictionary(model_path)

print dictionary
print corpus

# shuffle corpus
cp = list(corpus)
random.seed(326719) 
random.shuffle(cp)
```

```python
from pymongo import MongoClient

def dbClient():
    """
    This function makes a Connection with MongoClient, i.e., 
    creates a MongoClient to the running mongod instance. 
    """
    #Making a Connection with MongoClient
    client = MongoClient('localhost', 27017)
    #Getting a Database
    db = client['aolSearchDB']
    return db
```

```python
# split into 80% training and 20% test sets
#p = int(len(cp) * .9)
#cp = cp[p:]

p = int(len(cp) * .9)
cp_train = cp[0:p]
cp_test = cp[p:]


print "Corpus to Test:", len(cp_test)
print "Corpus to Train:", len(cp_train)
```

```python
import time
db = dbClient()

start_time = time.time()

alphas = [None, 0.1, 0.2, 0.3, 0.5, 0.6, 0.7, 0.8, 0.9, 1]
num_topics = [2, 5, 10, 50, 100, 150, 200]
times = []
perplexs = []
for a in alphas:    
    for nt in num_topics:        
        lda = gensim.models.ldamodel.LdaModel(corpus=cp_train, id2word=dictionary, num_topics=nt, alpha=a,
                                                  update_every=1, chunksize=2000, passes=1)

        elapsed = time.time() - start_time
        print lda
        print('Elapsed time: '),
        print elapsed
        print('Perplexity: '),
        perplex = lda.bound(cp_test)   
        pwp =  numpy.exp2(-perplex / sum(cnt for document in cp_test for _, cnt in document))
        perplexs.append((a, nt, perplex, pwp, elapsed))
                            
        values = {"alpha": a,
                 "topics" : nt, 
                 "perplex" : perplex, 
                 "per_word_perplexity" : pwp, 
                 "elapsed_time" : elapsed
        }   
        #Inserting Documents (AOl_Goals)
        lda_assessment = db.lda_assessment
        lda_assessment.insert(values)
        
        print perplexs
        print "------------------------------------------------------------------------------------------"
        print 
```
