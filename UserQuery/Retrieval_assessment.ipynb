```python
import pandas
def draw_ndcg(cursor, labels):

    fig = plt.figure(figsize=(24, 8))
    ax = fig.add_subplot(122)
    #colors = ['b', 'c', 'y', 'm', 'r', 'g']   
    colors = ['#b22222','#1e90ff','#0000cd','#00fa9a','#a020f0','#696969','#000000','#5f9ea0','#8b4513','#228b22','#ffd700','#483d8b']
    markers = ['^','.','o','.','^','s','<','>','v','h','H','d']
    lines = ['--', '-.', ':', '-', ':', '--', '-.', '-', ':',':','-','-.']
    i = 0    
    for cur in cursor:                     
        df_perp = pandas.DataFrame(cur)
 
        label = labels[i]
        #print label
        c =  ":" + str(markers[i]) 
        #c =  lines[i]  
        #c = '-x'
        plt.plot(df_perp, c, color=colors[i], label = label)  

        i = i + 1
        #if i == 6: i = 0
    legend = plt.legend(loc='best', shadow=True)
    plt.xlabel('Rank', fontsize=14)
    plt.ylabel('nDCG', fontsize=14)
    ax.set_ylim([0.4,1.05])
    ax.set_xlim([-1,50])
    plt.text(0.5, 1.04, 'Ganancia Acumulada Descontada Normalizada (nDCG)', fontsize=16, horizontalalignment='center', transform = ax.transAxes)
    plt.grid()
```
